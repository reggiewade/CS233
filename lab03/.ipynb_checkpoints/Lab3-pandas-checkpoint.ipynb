## Manipulating *DataFrame*s

<u>***Question 7. (pt 1)***</u> :

Create another *DataFrame* called `movies_df_2` containing only the columns `'Rank'` and `'Title'` in `movies_df`. 

```python
movies_df_2 = movies_df[['Rank', 'Title']]
movies_df_2
```

<u>***Question 8. (pt 1)***</u> :

Rename the column `'Votes'` in `movies_df` to `'Total Votes'`, and the column `'Title'` to `'Movie Title'`. 

```python
movies_df = movies_df.rename({'Votes': 'Total Votes', 'Title': 'Movie Title'}, axis="columns")
movies_df
```

<u>***Question 9. (pt 1)***</u> :

Create another *DataFrame* called `movies_df_3` that is `movies_df` with the column `'Rank'` removed. 

```python
movies_df_3 = movies_df[["Total Votes", "Movie Title", "Year", "Decade"]]
movies_df_3
```

<u>***Question 10. (pt 1)***</u> :

Sort `movies_df_3` by the values in the `'Decade'` and `'Movie Title'` columns. 

```python
movies_df_3 = movies_df_3.sort_values(by=['Decade', 'Movie Title'])
movies_df_3
```

<u>***Question 11. (pt 1)***</u> :

Reset the index in `movies_df_3`. Be sure the current index isn't included as a new column.

```python
movies_df_3 = movies_df_3.reset_index(drop=True)
movies_df_3
```

<u>***Question 12. (pt 1)***</u> :

Remove all of the columns with missing values in the *DataFrame* `new_df`. Don't save the changes.

```python
# reindex with more rows
new_df = movies_df_2.reindex(index=np.arange(1, 255, 1))  
# new rows have missing values 
new_df.tail(10) 
```

```python
new_df.dropna(axis=1, inplace=False)
```

<u>***Question 13. (pt 1)***</u> :

Remove all of the rows with missing values in the *DataFrame* `new_df`. This time save the changes.  

```python
new_df = new_df.dropna(axis=0, inplace=False)
new_df.tail(10)
```

<u>***Question 14. (pt 1)***</u> :

Create a new *DataFrame* called `movies_50s_60s` with all of the information about the movies from the 1950s and 1960s. 

```python
movies_50s_60s = (movies_df_3.loc[(movies_df_3['Decade'] >= 1950) & (movies_df_3['Decade'] <= 1960)])
movies_50s_60s.reset_index(drop=True)
```

<u>***Question 15. (pt 1)***</u> :

How many movies are there in each decade in the dataset? 

```python
print(f"Number of movies in 50s: {(movies_50s_60s['Decade'] == 1950).sum()}")
print(f"Number of movies in 60s: {(movies_50s_60s['Decade'] == 1960).sum()}")
```

## Using Functions on *DataFrame*s

<u>***Question 20. (pt 1)***</u> :

Apply the `reformat_votes` function below to the `Total Votes` column in `movies_df`?

```python
def reformat_votes(v):
    return v / 1000
```

```python
movies_df["Reformatted Votes"] = movies_df["Total Votes"].apply(reformat_votes)
movies_df
```

<u>***Question 21. (pt 1)***</u> :

Apply the `votes_per_rank` function below to the `Total Votes` and `Rank` values in each row of `movies_df`.

```python
def votes_per_rank(row):
    return row['Total Votes'] / row['Rank']
```

```python
movies_df["Votes Per Rank"] = movies_df.apply(votes_per_rank, axis=1)
movies_df
```

<u>***Question 22. (pt 1)***</u> :

Apply the `multiply_by_10` function below to each value in the `random_vals` *DataFrame*.

```python
def multiply_by_10(val):
    return val * 10

random_vals = pd.DataFrame(np.random.randint(0,100,size=(4, 3)), columns=list('ABC'))
```

```python
print(random_vals)
random_vals[['A', 'B', 'C']].apply(multiply_by_10)
```
