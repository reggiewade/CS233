### 1d. (4 pts) Create a query that retrieves for the years 1998 to 2001, inclusive, the species, genus, plot type, and year of each survey. 

```python
# 2d. Create a query 
query = '''
SELECT species.species AS "Species", 
       species.genus AS "Genus",
       plots.plot_type AS "Plot Type",
       surveys.year AS "Year"
FROM surveys INNER JOIN species
ON species.species_id=surveys.species_id
INNER JOIN plots
ON plots.plot_id = surveys.plot_id
WHERE (surveys.year >= 1998) AND (surveys.year <= 2001)
'''

cur.execute(query)
```

```python
# Display results 
results = cur.fetchall()
results
```

```python
# Create query to count number of entries (could also get the size of the list using len())
query = '''
SELECT COUNT(surveys.record_id) AS "Count" FROM surveys
WHERE (surveys.year >= 1998) AND (surveys.year <= 2001)
'''

cur.execute(query)
result = cur.fetchone()
num_entries = result[0]
```

### 6b. (4 pts) For all of the entries in the `surveys` table with the `plot_id` set to `10` and the year set to `1998`, change the `plot_id` to `25`. 

```python
# To change the plot_id to 25:  

# To begin, load table into DataFrame
query = """
SELECT * 
FROM surveys
"""
df = run_query(query)
df1 = df[(df.plot_id==10) & (df.year==1998)]
# show top 5 rows
df1.head()
```

```python
# Many ways to do this. 

# First, get second table without entries. 
# not(plot_id==10 AND year==1998) same as 
# not(plot_id==10) OR not(year==1998)
df2 = df[~(df.plot_id==10) | ~(df.year==1998)]

# show top 5 rows
df2.head()
```

```python
# Second, pull out entries need to change
df = df[(df.plot_id==10) & (df.year==1998)]

# update plot_id
df.plot_id=25

# show top 5 rows
df.head()
```

```python
# Next, recombine two tables 
df = pd.concat([df2, df])
df.head()
```

```python
# Lastly, replace table in database
df.to_sql("surveys", db, if_exists='replace')
```

```python
# Look up entry to verify changes 
query = '''
SELECT * FROM surveys
WHERE plot_id == 25; 
'''
df = run_query(query)
df
```
