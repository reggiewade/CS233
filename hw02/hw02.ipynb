```python
#dependencies
import numpy as np
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
from sqlalchemy import text
```

## Part 1: Accessing Data - Sqlite

```python
#create query
query = """
SELECT *
FROM species
"""
#fetch data
conn = sqlite3.connect("survey_data.sqlite")
cursor = conn.execute(query)
all_species = cursor.fetchall()
#release references
cursor.close()
#print data
for row in all_species:
    print(row)
```

```python
query = """
SELECT *
FROM plots
WHERE plot_type = "Control"
"""
cursor = conn.execute(query)
plots_control = cursor.fetchall()
cursor.close()

for row in plots_control:
    print(row)
```

```python
query = """
SELECT *
FROM species
WHERE taxa = "Bird"
"""

cursor = conn.execute(query)
first_taxa_bird = cursor.fetchone()
cursor.close()
print(first_taxa_bird)
```

```python
query = """
SELECT species, genus, plot_type, year
FROM surveys
INNER JOIN species ON surveys.species_id = species.species_id
INNER JOIN plots ON surveys.plot_id = plots.plot_id
WHERE year BETWEEN 1998 AND 2001
"""

cursor = conn.execute(query)
data_from_1998_to_2001 = cursor.fetchall()
cursor.close()

for row in data_from_1998_to_2001:
    print(row)
print(type(data_from_1998_to_2001))
```

```python
query = """
SELECT *
FROM species
WHERE taxa = "Bird"
"""

df_1b = pd.read_sql_query(query, conn)
df_1b
```

```python
query = """
SELECT species, genus, plot_type, year
FROM surveys
INNER JOIN species ON surveys.species_id = species.species_id
INNER JOIN plots ON surveys.plot_id = plots.plot_id
WHERE year BETWEEN 1998 AND 2001
"""

df_1c = pd.read_sql_query(query, conn)
df_1c
```

```python
query = """
SELECT year, COUNT(*) AS observation_count
FROM surveys
GROUP BY year;
"""

df_num_observations = pd.read_sql_query(query, conn)
df_num_observations.head(5)
```

```python
query = """
SELECT species, AVG(surveys.hindfoot_length) AS avg_hindfoot_length
FROM surveys
INNER JOIN species ON surveys.species_id = species.species_id
WHERE surveys.hindfoot_length IS NOT NULL
GROUP BY species.species;
"""

df_avg_hindfoot_length_species = pd.read_sql_query(query, conn)
df_avg_hindfoot_length_species.head(5)
```

## Part 3: Creating Tables - Sqlite 

```python
#create database survey_2002
query = """
    DROP TABLE IF EXISTS survey_2002
"""

conn.execute(query)

query = """
CREATE TABLE survey_2002 AS SELECT *
FROM surveys
WHERE year = 2002
"""

conn.execute(query)

#read survey data
query = """
SELECT *
FROM survey_2002
"""

#print data
survey02 = pd.read_sql_query(query, conn)
survey02.head(5)
```

## Part 4: Creating Tables - *pandas* 

```python
query = '''
SELECT * FROM surveys 
INNER JOIN species
ON surveys.species_id = species.species_id
WHERE species.taxa=='Reptile'
'''

df_reptile_data = pd.read_sql_query(query, conn)
df_reptile_data.head(5)
```

```python
#save species_id column
species_id = df_reptile_data.iloc[:, 5]
print(species_id.head(5))
```

```python
# Now drop extra data 
df_reptile_data.drop(['species_id', 'genus', 'species','taxa'], axis=1, inplace=True)
df_reptile_data.head()
```

```python
df_reptile_data['species_id'] = species_id
df_reptile_data
```

```python
df_reptile_data.to_sql("rep_table", conn)
```
